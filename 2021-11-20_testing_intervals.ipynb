# Proof by Contradiction

Let's do a proof by contradiction

$f(x) = x^2 + 2x + 1$

We want to find a zero of this equation

Claim: $x=2$ is a zero of $f$; i.e., $f(2) = 0$


Proof: Suppose by way of contradiction that $f(2) = 0$

Then I can plug 2 into $f$ and the result will be 0

$f(2) = (2)^2 + 2(2) + 1 = 4 + 4 + 1 = 9 \neq 0$

Therefore, the claim is false: $f(2) \neq 0$.

Hypothesis testing is a generalization of proof by contradiction to situations where you cannot definitively show that something is false; you can only show that it is very unlikely.

# Practice Generating Hypotheses
Pharmaceutical companies need to test whether their newly developed vaccines decrease $p$: the probability that someone contracts COVID-19. They conduct an experiment with a control group and an experimental group. The experimental group receives the COVID vaccine, but the control group does not. Out of the first group (the control group), the proportion of subjects that contract COVID is given by $\hat{p}_1$. In the second group (the experimental group), the corresponding proportion is $\hat{p}_2$. $\hat{p}_1$ is an estimate of $p_1$, and $\hat{p}_2$ is an estimate of $p_2$.

- $H_0$: $p_1 = p_2$
- $H_A$: $p_1 > p_2$
- Type I Error: Probability of concluding the vaccine is effective when it is not
- Type II Error: Probability of concluding the vaccine is ineffective when it actually is effective

# Central Limit Theorem for Binomial Data (Not Covered in Session)

$X \sim \text{Binomial}(n, \pi)$

$P = X/n$

So $0 \leq P \leq 1$

By the Central Limit Theorem, $P \xrightarrow{d} Y \sim \text{Normal}(\pi, \frac{\pi (1-\pi)}{n})$

The above notation means that $P$ converges in distribution to a random variable with a mean of $\pi$ and a variance of $\frac{\pi (1-\pi)}{n}$.

So $\text{Pr}(P \leq p) \approx \text{Pr}(Y \leq p)$

$Y$ can be related to the standard normal as follows:

$Y = \pi + \sqrt{\frac{\pi (1-\pi)}{n}} Z$

So $\text{Pr}(Y \leq p) = \text{Pr}(\pi + \sqrt{\frac{\pi (1-\pi)}{n}} Z \leq p) = \text{Pr}\Big(Z \leq (p-\pi) \sqrt{\frac{n}{\pi (1-\pi)}}\Big)$
